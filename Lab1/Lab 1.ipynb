```python
student_detail = { 'student_id' : ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S11', 'S12'],
                 'name': ["Daniella Fenton", "Ryder Storey", "Bryce Jensen", "Ed Bernal", "Kwame Morin","Mark Smith", "Bob Love", " Bruce Ong", " Alex Sander", "Allie Chang"],
                 'cumulative_marks' : [200,210,190,222,199,201,214,193,230,183]}

exam_data = {"student_id" : ['S1','S2', "S3", "S4", "S6", "S7", "S8", "S9", "S10", "S11", "S12", "S13"],
            "exam_score" : [23,45,12,67,21,55,33,14,56,83, 88, 12]}
```

```python
## Exercise 1.0
student_detail = pd.DataFrame(student_detail, columns = ["student_id", "name", "cumulative_marks"])
exam_data = pd.DataFrame(exam_data, columns = ["student_id", "exam_score"])
```

```python
print(student_detail)
print(exam_data)
```

```python
## Exercise 1.1
student_detail = pd.merge(student_detail, exam_data, on ="student_id", how = "left")
print(student_detail)
```

```python
## Exercise 1.2
student_detail.fillna(student_detail['exam_score'].median(), inplace = True)
print(student_detail)
```

```python
## Exercise 1.3
student_detail['final_score'] = (student_detail['cumulative_marks']  + student_detail['exam_score'])/4
print(student_detail)
```

```python
## Exercise 1.4
student_detail['letter_grade'] = "D"
for i in range(len(student_detail)):
    if (student_detail.quantile(0.25)['final_score'] < student_detail['final_score'][i] <= student_detail.quantile(0.5)['final_score']):
        student_detail.loc[i,'letter_grade'] = "C"
    elif (student_detail.quantile(0.5)['final_score'] < student_detail['final_score'][i] <= student_detail.quantile(0.75)['final_score']):
        student_detail.loc[i,'letter_grade'] = "B"
    elif (student_detail.quantile(0.75)['final_score'] < student_detail['final_score'][i]):
        student_detail.loc[i,'letter_grade'] = "A"

student_detail
```
